```python
# ============================================
!pip -q install xgboost shap imbalanced-learn

import os, json, textwrap, nbformat, warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from pathlib import Path
from datetime import datetime

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import (roc_auc_score, average_precision_score, f1_score, precision_score,
                             recall_score, confusion_matrix, classification_report, RocCurveDisplay,
                             PrecisionRecallDisplay)
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import mutual_info_classif
from sklearn.inspection import permutation_importance

from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import SMOTE

import shap
import xgboost as xgb
import joblib
```

```python
# 1) Configuraci√≥n y utilidades
OUTPUT_DIR = Path("/content/outputs")
OUTPUT_DIR.mkdir(exist_ok=True, parents=True)

RANDOM_STATE = 42
N_SPLITS = 5

def savefig(path):
    plt.tight_layout()
    plt.savefig(path, bbox_inches="tight", dpi=140)
    print(f"Figura guardada en: {path}")

def print_box(title):
    print("\n" + "="*80)
    print(title)
    print("="*80 + "\n")
```

```python
# 3) Carga de datos (USANDO TU datos_final.csv)
DATA_PATH = "/content/datos_final.csv"
assert os.path.exists(DATA_PATH), "No se encontr√≥ 'datos_final.csv' en /content"

df = pd.read_csv(DATA_PATH)
print_box("Vista general de los datos")
display(df.head(3))
print(df.shape, df.dtypes.value_counts())
```

```python
target_candidates = [c for c in df.columns if c.lower() in
                     ["churn","cancel","cancellation","baja","churn_flag","is_churn","cancelled","target", "cancelaci√≥n"]]
if len(target_candidates) == 0:
    raise ValueError("No se detect√≥ columna objetivo autom√°ticamente. Renombra tu y/target a 'churn' o ajusta 'target_col' abajo.")
target_col = target_candidates[0]
print(f"Columna objetivo detectada: {target_col}")
```

```python
# Convertir target a binario si viene como string/categ√≥rico
if df[target_col].dtype == 'O':
    df[target_col] = df[target_col].str.strip().str.lower().map({
        'si':1,'s√≠':1,'yes':1,'y':1,'true':1,'t':1,'1':1,'churn':1,'cancel':1,'cancelled':1,'baja':1,
        'no':0,'false':0,'f':0,'0':0,'stay':0,'retained':0
    }).fillna(df[target_col]).astype(int)
```

```python
# Dropear duplicados exactos
df = df.drop_duplicates().reset_index(drop=True)
```

```python
# Separar X, y
y = df[target_col].astype(int)
X = df.drop(columns=[target_col])
```

```python
# Identificar tipos
num_cols = [c for c in X.columns if pd.api.types.is_numeric_dtype(X[c])]
cat_cols = [c for c in X.columns if c not in num_cols]
```

```python
print_box("Columnas num√©ricas / categ√≥ricas")
print(f"Num√©ricas ({len(num_cols)}): {num_cols[:20]}{'...' if len(num_cols)>20 else ''}")
print(f"Categ√≥ricas ({len(cat_cols)}): {cat_cols[:20]}{'...' if len(cat_cols)>20 else ''}")
```

```python
# Train / Test split (estratificado)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE, stratify=y
)
```

```python
# 5) EDA r√°pido y correlaciones
print_box("Distribuci√≥n de la variable objetivo")
print(y.value_counts(normalize=True).rename("proporci√≥n"))
plt.figure(figsize=(5,4))
sns.countplot(x=y)
plt.title("Distribuci√≥n del target")
savefig(OUTPUT_DIR/"target_distribution.png")
plt.close()
```

```python
# Matriz de correlaci√≥n (solo num√©ricas)
if len(num_cols) >= 2:
    corr = X_train[num_cols].corr()
    plt.figure(figsize=(10,8))
    sns.heatmap(corr, cmap="coolwarm", center=0)
    plt.title("Correlaci√≥n entre variables num√©ricas (train)")
    savefig(OUTPUT_DIR/"correlations_heatmap.png")
    plt.close()
```

```python
# 6) Selecci√≥n de variables: Mutual Information + Colinealidad
pre_MI = ColumnTransformer(
    transformers=[
        ("num", Pipeline(steps=[
            ("imputer", SimpleImputer(strategy="median")),
            ("scaler", StandardScaler())
        ]), num_cols),
        ("cat", Pipeline(steps=[
            ("imputer", SimpleImputer(strategy="most_frequent")),
            ("ohe", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
        ]), cat_cols)
    ],
    remainder="drop"
)
```

```python
Xt_MI = pre_MI.fit_transform(X_train)
mi = mutual_info_classif(Xt_MI, y_train, random_state=RANDOM_STATE)
```

```python
# Nombres de features expandidos (incluye OHE)
feature_names_num = num_cols
ohe = pre_MI.named_transformers_["cat"].named_steps["ohe"] if len(cat_cols)>0 else None
feature_names_cat = list(ohe.get_feature_names_out(cat_cols)) if ohe is not None else []
mi_features = feature_names_num + feature_names_cat
```

```python
mi_df = pd.DataFrame({"feature": mi_features, "mutual_info": mi}).sort_values("mutual_info", ascending=False)
top_k = min(40, len(mi_df))
mi_top = mi_df.head(top_k)
display(mi_top.head(15))
mi_top.to_csv(OUTPUT_DIR/"mutual_info_top.csv", index=False)
```

```python
plt.figure(figsize=(8, max(4, int(top_k*0.25))))
sns.barplot(data=mi_top, x="mutual_info", y="feature")
plt.title("Top caracter√≠sticas por Mutual Information")
savefig(OUTPUT_DIR/"mi_top.png")
plt.close()
```

```python
# Reducir colinealidad en num√©ricas (VIF simple por correlaci√≥n alta)
selected_num = num_cols.copy()
if len(num_cols) > 1:
    corr_thr = 0.92
    corr_matrix = X_train[num_cols].corr().abs()
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    to_drop = [column for column in upper.columns if any(upper[column] > corr_thr)]
    selected_num = [c for c in num_cols if c not in to_drop]
    print_box(f"Eliminadas por alta correlaci√≥n (> {corr_thr}): {to_drop}")
```

```python
# Quedarnos con top_k MI + num no colineales
selected_from_mi = set(mi_top["feature"])
```

```python
# Para num√©ricas originales, si fueron eliminadas por colinealidad y aparecen como nombres originales:
selected_from_mi = {f for f in selected_from_mi if (f in feature_names_cat) or (f in selected_num)}
```

```python
# 7) Preprocesamiento final (imputaci√≥n, OHE, escalado)
def default_preprocess(num_cols, cat_cols):
    return ColumnTransformer(
        transformers=[
            ("num", Pipeline(steps=[
                ("imputer", SimpleImputer(strategy="median")),
                ("scaler", StandardScaler()),
            ]), num_cols),
            ("cat", Pipeline(steps=[
                ("imputer", SimpleImputer(strategy="most_frequent")),
                ("ohe", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
            ]), cat_cols),
        ],
        remainder="drop"
    )
```

```python
preprocess = default_preprocess(num_cols, cat_cols)
if "custom_preprocess" in globals() and callable(globals()["custom_preprocess"]):
    print("üß© Usando 'custom_preprocess' de tus notebooks.")
    preprocess = globals()["custom_preprocess"](num_cols, cat_cols)
```

```python
# 8) Definici√≥n de modelos
models = {
    "LogReg": LogisticRegression(max_iter=200, class_weight="balanced", random_state=RANDOM_STATE),
    "RandomForest": RandomForestClassifier(
        n_estimators=400, max_depth=None, min_samples_split=2,
        n_jobs=-1, random_state=RANDOM_STATE, class_weight="balanced_subsample"
    ),
    "XGBoost": xgb.XGBClassifier(
        n_estimators=500, max_depth=6, learning_rate=0.05, subsample=0.9, colsample_bytree=0.9,
        reg_lambda=1.0, objective="binary:logistic", tree_method="hist", random_state=RANDOM_STATE,
        eval_metric="auc"
    )
}
```

```python
# Manejo de desbalance con SMOTE dentro del pipeline
def make_pipeline(clf):
    return ImbPipeline(steps=[
        ("preprocess", preprocess),
        ("smote", SMOTE(random_state=RANDOM_STATE, k_neighbors=5)),
        ("clf", clf)
    ])
```

```python
# 9) Entrenamiento + Validaci√≥n cruzada y evaluaci√≥n en test

results = []
trained = {}

for name, clf in models.items():
    print_box(f"Entrenando modelo: {name}")
    pipe = make_pipeline(clf)

    # CV ROC AUC
    cv = StratifiedKFold(n_splits=N_SPLITS, shuffle=True, random_state=RANDOM_STATE)
    cv_auc = cross_val_score(pipe, X_train, y_train, scoring="roc_auc", cv=cv, n_jobs=-1)
    print(f"AUC-ROC CV ({N_SPLITS} folds): {cv_auc.mean():.4f} ¬± {cv_auc.std():.4f}")

    # Train and evaluate on test set
    pipe.fit(X_train, y_train)
    trained[name] = pipe # Store trained pipeline

    y_pred = pipe.predict(X_test)
    y_proba = pipe.predict_proba(X_test)[:, 1]

    # Metrics
    auc_test = roc_auc_score(y_test, y_proba)
    ap_test = average_precision_score(y_test, y_proba)
    f1_test = f1_score(y_test, y_pred)
    precision_test = precision_score(y_test, y_pred)
    recall_test = recall_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)

    print(f"AUC-ROC Test: {auc_test:.4f}")
    print(f"Avg Precision Test: {ap_test:.4f}")
    print(f"F1 Score Test: {f1_test:.4f}")
    print(f"Precision Test: {precision_test:.4f}")
    print(f"Recall Test: {recall_test:.4f}")
    print("Confusion Matrix:\n", cm)
    print("\nClassification Report:\n", classification_report(y_test, y_pred))

    results.append({
        "model": name,
        "auc_cv_mean": cv_auc.mean(),
        "auc_cv_std": cv_auc.std(),
        "auc_test": auc_test,
        "ap_test": ap_test,
        "f1_test": f1_test,
        "precision_test": precision_test,
        "recall_test": recall_test,
    })
```

```python
print_box("Resultados de los modelos")
results_df = pd.DataFrame(results).sort_values("auc_test", ascending=False)
display(results_df)
results_df.to_csv(OUTPUT_DIR/"model_performance_summary.csv", index=False)
```

```python
# 10) Visualizaci√≥n de resultados
print_box("Curvas ROC y Precision-Recall")
plt.figure(figsize=(12, 5))

ax1 = plt.subplot(1, 2, 1)
for name, pipe in trained.items():
    RocCurveDisplay.from_estimator(pipe, X_test, y_test, ax=ax1, name=name)
ax1.set_title("Curvas ROC")

ax2 = plt.subplot(1, 2, 2)
for name, pipe in trained.items():
    PrecisionRecallDisplay.from_estimator(pipe, X_test, y_test, ax=ax2, name=name)
ax2.set_title("Curvas Precision-Recall")

savefig(OUTPUT_DIR/"roc_pr_curves.png")
plt.show()
```

```python
# Obtener nombres de features finales del ColumnTransformer
def get_feature_names(column_transformer):
    feature_names = []
    for name, trans, cols in column_transformer.transformers_:
        if name == "remainder" and trans == "drop":
            continue
        if hasattr(trans, "named_steps"):
            # Pipeline
            last = list(trans.named_steps.values())[-1]
            if hasattr(last, "get_feature_names_out"):
                try:
                    fn = list(last.get_feature_names_out(cols))
                except:
                    fn = cols
            else:
                fn = cols
        else:
            if hasattr(trans, "get_feature_names_out"):
                try:
                    fn = list(trans.get_feature_names_out(cols))
                except:
                    fn = cols
            else:
                fn = cols
        feature_names.extend(list(fn))
    return feature_names

final_features = get_feature_names(trained["LogReg"].named_steps["preprocess"])
```


# üìå Conclusiones Estrat√©gicas del Modelo de Cancelaci√≥n


### Interpretaci√≥n Estrat√©gica
- Estos factores deben ser el foco principal de cualquier estrategia de **retenci√≥n de clientes**.
- Un aumento o deterioro en estas variables incrementa la probabilidad de cancelaci√≥n, por lo que son **indicadores clave de riesgo**.
- La empresa deber√≠a:
  - Implementar **alertas tempranas** cuando los clientes muestren se√±ales en estas variables.
  - Definir **campa√±as espec√≠ficas** para clientes de alto riesgo (ej. descuentos, planes personalizados).
  - Reentrenar el modelo peri√≥dicamente para capturar cambios en el comportamiento.

### Recomendaciones Adicionales
- Ajustar el umbral de decisi√≥n seg√∫n la estrategia:  
  - Si el costo de perder un cliente es alto, conviene priorizar **Recall** (identificar la mayor cantidad de posibles bajas, aunque aumenten falsos positivos).  
  - Si el costo de intervenir es alto, priorizar **Precisi√≥n** (centrarse en clientes con mayor probabilidad real de cancelar).  
- Implementar un **monitoreo de drift** para detectar si la distribuci√≥n de las variables cambia con el tiempo.
- Seguir enriqueciendo el dataset con variables de uso, interacciones con soporte y engagement digital, para mejorar el poder predictivo.
